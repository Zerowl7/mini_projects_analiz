## Приходят бренды и хотят сделать скидку лояльным клиентам 

### tc, art_sp

```python
user_df = df[['tc', 'art_sp']]
```

```python
user_df = user_df.rename(columns ={'tc': 'user_id',
                                  'art_sp': 'brand'
                                  })
```

```python
user_df
```

### достаём информацию из бренда  -->  Store_Brand

```python
brand_name = 'MARAVILLA 500 G Store_Brand'
```

```python
brand_name.split(' ')[-1]
```

### Создаём новую колонку

```python
user_df['brand_name'] = user_df.brand.apply(lambda x: x.split(' ')[-1])
```

```python
user_df
```

# Разведка

### Гипотеза о пользователях n окупок для бренда 

```python

users_purchases = user_df.groupby('user_id', as_index = False)\
    .agg({'brand_name': "count"})\
    .rename(columns={'brand_name': 'purchases'})\
    .query('purchases >= 5')
# количество покупок больше, чем 5 (75%)
```

```python
users_unique_brands = user_df.groupby('user_id', as_index = False)\
        .agg({'brand_name': pd.Series.nunique})\
        .rename(columns = {'brand_name': 'unique_brands'})
# количество уникальных брендов
```

## Считаем, какой процент будет составлять любимого бренда клиента от общего процента покупок

```python
lovely_brand_purchases_df = user_df.groupby(['user_id', 'brand_name'], as_index = False)\
        .agg({'brand': 'count'})\
        .sort_values(['user_id', 'brand'], ascending = [False, False])\
        .groupby('user_id')\
        .head(1)\
        .rename(columns={'brand_name': 'lovely_brand',
                        'brand': 'lovely_brand_purcheses'})
    #head очень интересно здесь используется, так как они все отсортировано
```

```python
lovely_brand_purchases_df
```

# Объеденяем информацию в один датасет

```python
loyalty_df = users_purchases.merge(users_unique_brands, on='user_id')\
               .merge(lovely_brand_purchases_df, on='user_id')
```

```python
loyalty_df
```

```python
loyal_users = loyalty_df[loyalty_df.unique_brands == 1]
```

```python
loyalty_df['loyalty_score'] = loyalty_df.lovely_brand_purcheses / loyalty_df.purchases
```
